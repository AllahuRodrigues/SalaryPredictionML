```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("survey_results_public.csv")

```

```python
df.head()
```

```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]
df = df.rename(columns={"ConvertedCompYearly": "Salary"})
df.head()
```

```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()

print(df.columns)
```

```python
df = df[df["Employment"].str.contains("Employed", case=False, na=False)]  # Keep only employed individuals
df = df.drop("Employment", axis=1)  # Now it's safe to drop the column

df.info()

print(df.shape)  # Should show a non-zero number of rows
print(df.head())  # Preview the filtered dataset
```

```python
df['Country'].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```
